```python
import os, random, shutil

selected_classes = [
  "Tomato_healthy",
  "Tomato_Late_blight",
  "Potato___healthy",
  "Potato___Early_blight"
]
N_PER_CLASS = 300   # keep small for a fast demo; reduce to 150 if too slow

dst = "/content/mini_plant"   # new mini dataset
# remove old if present (be careful)
if os.path.exists(dst):
    shutil.rmtree(dst)
for split in ("train","val","test"):
    for cls in selected_classes:
        os.makedirs(os.path.join(dst, split, cls), exist_ok=True)

for cls in selected_classes:
    src_cls = os.path.join(data_dir, cls)
    imgs = [f for f in os.listdir(src_cls) if f.lower().endswith(('.jpg','.png','.jpeg'))]
    random.shuffle(imgs)
    imgs = imgs[:N_PER_CLASS]
    n_train = int(0.7 * len(imgs))
    n_val = int(0.15 * len(imgs))
    n_test = len(imgs) - n_train - n_val
    for i, im in enumerate(imgs):
        if i < n_train:
            split = "train"
        elif i < n_train + n_val:
            split = "val"
        else:
            split = "test"
        shutil.copy(os.path.join(src_cls, im), os.path.join(dst, split, cls, im))

print("Mini dataset ready at:", dst)
# show counts
for split in ("train","val","test"):
    print(split, {cls: len(os.listdir(os.path.join(dst, split, cls))) for cls in selected_classes})
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = (224, 224)
BATCH = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_directory(
    "/content/mini_plant/train",
    target_size=IMG_SIZE,
    batch_size=BATCH,
    classes=selected_classes,
    class_mode='categorical'
)

val_gen = val_datagen.flow_from_directory(
    "/content/mini_plant/val",
    target_size=IMG_SIZE,
    batch_size=BATCH,
    classes=selected_classes,
    class_mode='categorical',
    shuffle=False
)

test_gen = val_datagen.flow_from_directory(
    "/content/mini_plant/test",
    target_size=IMG_SIZE,
    batch_size=BATCH,
    classes=selected_classes,
    class_mode='categorical',
    shuffle=False
)
```
